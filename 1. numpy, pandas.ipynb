```python
# 1. numpy->데이터 분석 패키지
# 넘파이는 변수명이 없음
#!pip install numpy

import numpy as np

# np.array(리스트 형태로 입력)
v1 = np.array([1,2,3,4])
print(v1)
```

```python
# np.arange : 배열 생성
v1=np.arange(10)
print(v1)

# 데이터type은 생략가능
v2=np.arange(1,10,2, dtype=int)
v2=np.arange(3.5, 10.5, 2, dtype=float)
print(v2)
```

```python
# 제곱근 생성
v4=np.arange(1, 10, 2)**2
print(v4)

# 세제곱근 생성
v5=np.arange(1, 10, 2)**3
print(v5)
```

```python
# np.reshape : 배열->다른형태로 변환
# 1) 1차원 배열생성
v1 = np.arange(12)
print(v1)

# 2) 2차원 배열로 변환
v2=v1.reshape(2,6)
print(v2)

v3=v1.reshape(2,6,order="F")
print(v3)

v4=v1.reshape(2,6,order="C")
print(v4)
```

```python
# 행렬의 계산
v1=np.arange(1,5).reshape(2,2)
print(v1)

# 덧셈
print(np.add(v1,v1))
# 뺄셈
print(np.subtract(v1,v1))
# 곱셈
print(np.multiply(v1,v1))
# 내적
print(np.dot(v1,v1))
```

```python
# 최대, 최소값 => np.amax(), np.amin()

# 1) 1차원 배열 생성
v3=np.arange(3.5, 10.5, 2, dtype=float)
print(v3)

# 2) 1차원 배열 생성 후 2차원 배열로 변환
v4=np.arange(1,5).reshape(2,2)
print(v4)

# 3) 최댓값, 최솟값 확인
print(np.amax(v3))
print(np.amin(v4))

# 데이터타입 확인
print(v3.dtype)
print(v4.dtype)
```

```python
# pandas를 이용하여 복사, 추가, 삭제
# 데이터프레임명.copy() : 전체 카피
df_columns=df.copy()
df_columns

# 컬럼명 확인
df.columns

# 일부 열만 가져와서 새로운 데이터프레임 구성하고자 함
df_columns=df_columns[['size', 'period', 'price']]
print(df_columns.columns)

# 열 변수이름 변경
# 데이터프레임명.rename(columns={'변수명':'새로운 변수명'}, inplace=True)
# inplace=True : 덮어씌우라는 의미
df_columns.rename(columns={'period':'time'}, inplace=True)
print(df_columns.columns)

# 열끼리 계산을 통해 새로운 변수 생성
df_columns['growth']=df_columns['size']/df_columns['time']
df_columns.head()

# 변수 삭제
del df_columns['growth']
df_columns.head()
```

```python
# sql마냥 특정 케이스 추출
# 데이터프레임명[(데이터프레임명['변수명']==조건) & (데이터프레임명['변수명']==조건)]
# & : 두 조건다 충족해야함
# | : OR(둘중 하나)
df_continent_brand=df[(df['continent']==1) & (df['brand']==1)]
df_continent_brand

df_over_size_period=df[(df['size']>=10) & (df['period']>=30)]
df_over_size_period.head()
```

```python
# 값을 새로운 범주의 값으로 재할당할 때
# 데이터셋명['변수명'].value_counts() : 범주의 수를 확인한다
df['brand'].value_counts()

#2    24
#1    23
#3    16

# 방법1) replace 함수를 사용
recode_brand = {"brand" : {1:1, 2:1, 3:2}}
df_recode1=df.replace(recode_brand)
df_recode1.head()

df_recode1['brand'].value_counts()
#1    47
#2    16

# 방법2) 함수 정의
def brand_groups(series):
    if series==1:
        return 1
    elif series==2:
        return 1
    elif series==3:
        return 2

# brand_groups 함수에 df['brand']변수를 적용시켜 df['re_brand']라는 새로운 변수를 생성한다.
df['re_brand']=df['brand'].apply(brand_groups)
df.head()    
```

```python
# pandas와 numpy의 전환
# 판다스(기본분석) -> 넘파이(빠른 분석)
# 넘파이(눈으로 데이터 확인어렵) -> 판다스(데이터 형태 확인)

import pandas as pd
df=pd.read_csv('EX_GrapeData.csv')
print(df.head())

# numpy로 변환, 변수명이 없음! 
# 데이터프레임명.to_numpy()
df_num=df.to_numpy()
print(df_num)

# pandas로 변환, 변수명은 0,1,2,3...처럼 인덱스로 표현됨
# pd.DataFrame(데이터프레임명)
df_pd=pd.DataFrame(df_num)
print(df_pd)

# pandas로 변환 + 컬럼명 부여
df_pd2=pd.DataFrame(data=df_num, columns=['continet', 'brand', 'size', 'period', 'price'])
df_pd2.head()
```
